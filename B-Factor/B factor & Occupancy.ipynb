```python
from biopandas.pdb import PandasPdb
import pandas as pd
```

```python
# Open the file
file = PandasPdb().read_pdb('1OKE.pdb')
```

```python
data = file.df['ATOM']
data
```

```python
alpha = data.loc[data["atom_name"]=="CA"]
alpha
```

```python
len(alpha)
```

```python
chainA = alpha.loc[data["chain_id"]=="A"][["residue_number","b_factor"]]
chainA = chainA.rename({'b_factor': 'Chain A'}, axis=1)
chainA
```

```python
chainB = alpha.loc[data["chain_id"]=="B"][["residue_number","b_factor"]]
chainB = chainB.rename({'b_factor': 'Chain B'}, axis=1)
chainB
```

```python
bfact = pd.merge(left=chainA, right=chainB, left_on='residue_number', right_on='residue_number')
bfact["Average"] = (bfact["Chain A"]+bfact["Chain B"])/2
bfact
```

```python
# save the dataframe as a csv file

bfact.to_csv(r'b factor.txt', sep="\t", index=False)
```

```python
bfact_Average = bfact[["residue_number","Average"]]
bfact_Average
```

```python
# save the dataframe as a csv file

bfact_Average.to_csv(r'b factor averages.txt', sep="\t", index=False)
```

```python
occupancy = alpha[["residue_number","occupancy"]]
occupancy
```

```python
# save the dataframe as a csv file

occupancy.to_csv(r'occupancy.txt', sep="\t", index=False)
```
