```python
import pandas as pd
```

```python
# Open the file
full = pd.read_table("1oke RMSD protein.txt", sep=' ', header=None)
full = full[[0,1]]
full.columns = ['Residue ID', '1okeRMSD']
full
```

```python
# Open the file
dm1 = pd.read_table("DM1-1OKE.txt", sep=' ', header=None)
dm1 = dm1[[0,1]]
dm1.columns = ['Residue ID', 'dm1RMSD']
dm1
```

```python
# Open the file
dm2 = pd.read_table("DM2-1OKE.txt", sep=' ', header=None)
dm2 = dm2[[0,1]]
dm2.columns = ['Residue ID', 'dm2RMSD']
dm2
```

```python
# Open the file
dm3 = pd.read_table("DM3-1OKE.txt", sep=' ', header=None)
dm3 = dm3[[0,1]]
dm3.columns = ['Residue ID', 'dm3RMSD']
dm3
```

```python
rmsd = full.merge(dm1, on="Residue ID").merge(dm2,on="Residue ID").merge(dm3,on="Residue ID")
rmsd
```

```python
rmsd.mean()
```

```python
rmsd.std()
```
