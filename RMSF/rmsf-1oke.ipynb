```python
import pandas as pd
```

```python
# Open the file
CA = pd.read_table("rmsf_Averages.txt", sep='\t', header=None)
CA = CA[[0,1]]
CA.columns = ['Residue ID', 'CArmsf']
CA
```

```python
# Open the file
backbone = pd.read_table("rmsf_backbone_Averages.txt", sep='\t', header=None)
backbone = backbone[[0,1]]
backbone.columns = ['Residue ID', 'backboneRMSF']
backbone
```

```python
# Open the file
sidechain = pd.read_table("rmsf_sidechain_Averages.txt", sep='\t', header=None)
sidechain = sidechain[[0,1]]
sidechain.columns = ['Residue ID', 'sidechainRMSF']
sidechain
```

```python
rmsf = CA.merge(backbone, on="Residue ID").merge(sidechain,on="Residue ID")
rmsf
```

```python
rmsf.mean()
```

```python
rmsf.std()
```
