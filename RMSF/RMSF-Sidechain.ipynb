```python
import pandas as pd
import numpy as np
```

```python
# Open the file
data = pd.read_table("RMSF_sidechain.txt", sep=' ', header=None)
data
```

```python
data = data[[0,1,3]]
data.columns = ['Chain_ID','Residue_ID','RMSF']
data
```

```python
chainA = data.loc[data["Chain_ID"]==1][["Residue_ID","RMSF"]]
chainA
```

```python
chainB = data.loc[data["Chain_ID"]==0][["Residue_ID","RMSF"]]
chainB
```

```python
chainA_summary = pd.DataFrame(columns=['Residue', 'Average RMSF A'])
for i in chainA.Residue_ID.unique():
    avg = chainA.loc[chainA["Residue_ID"] == i]["RMSF"].mean()    
    chainA_summary.loc[i] = [i,avg]
chainA_summary
```

```python
chainB_summary = pd.DataFrame(columns=['Residue', 'Average RMSF B'])
for i in chainB.Residue_ID.unique():
    avg = chainB.loc[chainB["Residue_ID"] == i]["RMSF"].mean()    
    chainB_summary.loc[i] = [i,avg]
chainB_summary
```

```python
final = pd.merge(left=chainA_summary, right=chainB_summary, left_on='Residue', right_on='Residue')
final["Average"] = (final["Average RMSF A"]+final["Average RMSF B"])/2
final
```

```python
sidechain = final[["Residue","Average"]]
sidechain
```

```python
data_types_dict = {'Residue': int}
sidechain = sidechain.astype(data_types_dict)
sidechain
```

```python
# save the dataframe as a csv file

sidechain.to_csv(r'rmsf_sidechain_Averages.txt', sep="\t", index=False)
```
