```python
import pandas as pd
```

```python
# Open the file
data = pd.read_table("RMSF_backbone2.txt", sep=' ', header=None)
data
```

```python
data = data[[0,1,2]]
data.columns = ['Chain ID','ATOM','RMSF']
data
```

```python
chainA = data.loc[data["Chain ID"]==1][["ATOM","RMSF"]]
chainA
```

```python
chainA.iat[-2,1] = (chainA.iat[-1,1] + chainA.iat[-2,1]) / 2
chainA=chainA.drop(chainA.index[-1])
chainA
```

```python
chainB = data.loc[data["Chain ID"]==0][["ATOM","RMSF"]]
chainB
```

```python
chainB.iat[-2,1] = (chainB.iat[-1,1] + chainB.iat[-2,1]) / 2
chainB=chainB.drop(chainB.index[-1])
chainB
```

```python
chainA_summary = pd.DataFrame(columns=['Residue', 'Average RMSF A'])
residue = 1
for i in range(0,len(chainA),4):
    avg = chainA.iloc[i:i+4,:]["RMSF"].mean()    
    chainA_summary.loc[residue] = [residue,avg]
    residue+=1
chainA_summary
```

```python
chainB_summary = pd.DataFrame(columns=['Residue', 'Average RMSF B'])
residue = 1
for i in range(0,len(chainB),4):
    avg = chainB.iloc[i:i+4,:]["RMSF"].mean()    
    chainB_summary.loc[residue] = [residue,avg]
    residue+=1
chainB_summary
```

```python
final = pd.merge(left=chainA_summary, right=chainB_summary, left_on='Residue', right_on='Residue')
final["Average"] = (final["Average RMSF A"]+final["Average RMSF B"])/2
final
```

```python
backbone = final[["Residue","Average"]]
backbone
```

```python
data_types_dict = {'Residue': int}
backbone = backbone.astype(data_types_dict)
backbone
```

```python
# save the dataframe as a csv file

backbone.to_csv(r'rmsf_backbone_Averages.txt', sep="\t", index=False)
```
