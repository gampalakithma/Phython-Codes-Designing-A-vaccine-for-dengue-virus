```python
import pandas as pd
```

```python
# Open the file
data = pd.read_table("RMSF.txt", sep=' ', header=None)
data = data[[0,1]]
data.columns = ['Residue ID', 'RMSF']
data
```

```python
chainA = data.iloc[:394,:]
chainA = chainA.rename({'RMSF': 'Chain A'}, axis=1)
chainA
```

```python
chainB = data.iloc[394:,:]
chainB = chainB.rename({'RMSF': 'Chain B'}, axis=1)
chainB["Residue ID"] = [i for i in range(1,395)]
chainB
```

```python
rmsf = pd.merge(left=chainA, right=chainB, left_on='Residue ID', right_on='Residue ID')
rmsf["Average"] = (rmsf["Chain A"]+rmsf["Chain B"])/2
rmsf
```

```python
# save the dataframe as a csv file

rmsf.to_csv(r'rmsf_result.txt', sep="\t", index=False)
```

```python
rmsf_Average = rmsf[["Residue ID","Average"]]
rmsf_Average
```

```python
# save the dataframe as a csv file

rmsf_Average.to_csv(r'rmsf_Averages.txt', sep="\t", index=False)
```
